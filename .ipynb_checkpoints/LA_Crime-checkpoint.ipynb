```python
# Import dependencies
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Read CSV file
la_file = "Resources/LA_Crime_Data.csv"
la_df = pd.read_csv(la_file, encoding= "unicode_escape")
```

```python
# Evaluate data
la_df.head()
la_df = la_df.dropna()
la_df.info()
```

```python
# Convert reported data to pandas datetime64 type and use DatetimeIndex to add year and month data
# NOTE:  First convert date from an object to datetime64 use info() method to confirm panda data type for Reporte Dates
la_df["Date"] = pd.to_datetime(la_df["Date"])
la_df["Year"] = pd.DatetimeIndex(la_df["Date"]).year
la_df["Month"] = pd.DatetimeIndex(la_df["Date"]).month
la_df.info()
```

```python
# Review offense types and count
la_df["Offense Type"].value_counts()
la_df.head()
```

```python
# Group data by offense type and organize 
la_df_grouped = la_df.groupby("Offense Type").count()
la_df_grouped = la_df.sort_values("Date", ascending=False)
la_df_grouped.head(10)
```

```python
# Create sunnnary tabel of Denver Crime Statistics
la_summary_df = la_df.groupby("Offense Type").count()
la_summary_df = la_summary_df.sort_values(by=["isViolent"], ascending=False)
la_summary_df = la_summary_df["isViolent"]

pd.DataFrame({"# Reported Offenses": la_summary_df}).head(10)
```

```python
# Pie chart data

# Violent vs. Non_Violent Crime
violent_crimes = la_df["isViolent"].loc[la_df["isViolent"] == "Violent"].count()
non_violent_crimes = la_df["isViolent"].loc[la_df["isViolent"] == "Non-violent"].count()

# Top 10 LA Crime dataframes
batt_df = la_df.loc[la_df["Offense Type"] == "BATTERY - SIMPLE ASSAULT"]
vehstole_df = la_df.loc[la_df["Offense Type"] == "VEHICLE - STOLEN"]
burgveh_df = la_df.loc[la_df["Offense Type"] == "BURGLARY FROM VEHICLE"]
theftplain_df = la_df.loc[la_df["Offense Type"] == "THEFT PLAIN - PETTY ($950 & UNDER)"]
vandalism_df = la_df.loc[la_df["Offense Type"] == "VANDALISM - FELONY ($400 & OVER, ALL CHURCH VANDALISMS)"]
burglary_df = la_df.loc[la_df["Offense Type"] == "BURGLARY"]
simpassault_df = la_df.loc[la_df["Offense Type"] == "INTIMATE PARTNER - SIMPLE ASSAULT"]
assault_df = la_df.loc[la_df["Offense Type"] == "ASSAULT WITH DEADLY WEAPON, AGGRAVATED ASSAULT"]
theftveh_df = la_df.loc[la_df["Offense Type"] == "THEFT FROM MOTOR VEHICLE - PETTY ($950 & UNDER)"]
robbery_df = la_df.loc[la_df["Offense Type"] == "ROBBERY"]

# Top 10 Crime data for pie chart
Total = la_df["Offense Type"].count()
batt = batt_df["Offense Type"].count()
vehstole = vehstole_df["Offense Type"].count()
burgveh = burgveh_df["Offense Type"].count()
theftplain = theftplain_df["Offense Type"].count()
vandalism = vandalism_df["Offense Type"].count()
burglary = burglary_df["Offense Type"].count()
simpassault = simpassault_df["Offense Type"].count()
assault = assault_df["Offense Type"].count()
theftveh = theftveh_df["Offense Type"].count()
robbery = robbery_df["Offense Type"].count()
OtherOther = (Total - batt - vehstole - burgveh - theftplain - vandalism - burglary - simpassault - 
              assault - theftveh - robbery)

print(non_violent_crimes, violent_crimes)
print(batt,vehstole,burgveh,theftplain,vandalism,burglary,simpassault, assault, theftveh, robbery)

OtherOther
Total
```

```python
# Plot pie chart of crime types

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
plt.suptitle("Los Angeles Crime Summary")

# First Plot
values1 = [non_violent_crimes, violent_crimes]
labels1 = ["Non-Violent", "Violent"]
ax1.pie(values1, labels=labels1, autopct="%1.1f%%", shadow=False, startangle=90)

# Second Plot
values2 = [batt,vehstole,burgveh,theftplain,vandalism,burglary,simpassault, assault, theftveh, robbery, OtherOther]
labels2 = ["BATTERY - SIMPLE", "VEHICLE - STOLEN", "BURGLARY FROM VEHICLE", "THEFT PLAIN", "VANDALISM", 
           "BURGLARY", "SIMPLE ASSAULT", "AGGRAVATED ASSAULT", "THEFT FROM MOTOR VEHICLE", "ROBBERY", 
           "Other Major Crimes"]
ax2.pie(values2, labels=labels2, autopct="%1.1f%%", shadow=False, startangle=180)

# Plot
plt.subplots_adjust(wspace=.5)
plt.savefig("LApie/fig1.png", dpi=300)   
plt.show()
```

```python
# Group data into Violent / Non-violent categories, trends by month 
la_trend_df = la_df.groupby(["isViolent", "Year", "Month"]).count()
# dtrend_df = dtrend_df.sort_values(by=["Year", "Month"], ascending=True)dddd
la_trend_df.head()
```

```python
# Break trend data into violent and non-violent data frames and sort by Year and Month
# Nonviolent data
nonviolent_df = la_trend_df.loc["Non-violent"]
nonviolent_df = nonviolent_df.sort_values(by=["Year", "Month"], ascending=True)

# Violent data
violent_df = la_trend_df.loc["Violent"]
violent_df = violent_df.sort_values(by=["Year", "Month"], ascending=True)
```

```python
# Generate line plot as figure
# Step 1: Configure chart, xticks, and labels/titles
plt.figure(figsize=(12, 5))
plt.xticks(np.arange(16), ["Mar-19", "Apr-19", "May-19", "June-19", "Jul-19", "Aug-19", "Sep-19", "Oct-19", 
                      "Nov-19", "Dec-19", "Jan-20", "Feb-20", "Mar-20", "Apr-20", "May-20", "Jun-20"])
plt.title("LA City Monthly Crime Rates -- Violent vs. Non-Violent")
plt.xlabel("Month")
plt.ylabel("# Reported Cases")

# Step 2:  Configure data and plot
plt.plot(list(nonviolent_df["Offense Type"]), label="Non Violent Crimes")
plt.plot(list(violent_df["Offense Type"]), label="Violent Crimes")
plt.legend(loc="center right")
plt.savefig("Output Folder/latrend.png", dpi=300) 
plt.show()
```
