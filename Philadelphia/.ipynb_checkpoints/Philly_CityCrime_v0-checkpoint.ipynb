```python
# Import dependencies
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Read CSV files
philly_file = "Input/Philadelphia.csv"
philly_df = pd.read_csv(philly_file, encoding= "unicode_escape")
```

```python
# Evaluate data, drop na
philly_df.head()
philly_df = philly_df.dropna()
philly_df.info()
```

```python
# Convert reported data to pandas datetime64 type and use DatetimeIndex to add year and month data
# NOTE:  First convert date from an object to datetime64 use info() method to confirm panda data type for Reporte Dates
philly_df["Date"] = pd.to_datetime(philly_df["Date"])
philly_df["Year"] = pd.DatetimeIndex(philly_df["Date"]).year
philly_df["Month"] = pd.DatetimeIndex(philly_df["Date"]).month
philly_df.info()
```

```python
# Review offense types and clean data (remove traffic violations)
philly_df["Offense"].value_counts()
philly_clean_df = philly_df.loc[philly_df["Offense"] != "traffic-accident"]
philly_clean_df["Offense"].value_counts()
philly_clean_df.head()
```

```python
# Group data by offense type and organize 
philly_offense_df = philly_clean_df.groupby("Offense").count()
philly_offense_df = philly_offense_df.sort_values("Date", ascending=False)
philly_offense_df.head(10)
```

```python
# Create summary table of Philadelphia Crime Statistics
philly_crime_summary_df = philly_clean_df.groupby("Offense").count()
philly_crime_summary_df = philly_crime_summary_df.sort_values(by=["Is Violent"], ascending=False)
philly_crime_summary_df = philly_crime_summary_df["Is Violent"]

pd.DataFrame({"# Reported Offenses": philly_crime_summary_df})
```

```python
# Pie chart data

# Violent vs. Non_Violent Crime
violent_crimes = philly_clean_df["Is Violent"].loc[philly_clean_df["Is Violent"] == "Violent"].count()
non_violent_crimes = philly_clean_df["Is Violent"].loc[philly_clean_df["Is Violent"] == "Non-Violent"].count()

# Top 10 Philadelphia Crime dataframes
AllOther_df = philly_clean_df.loc[philly_clean_df["Offense"] == "All Other Offenses"]
Thefts_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Thefts"]
Other_Assaults_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Other Assaults"]
Vandalism_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Vandalism/Criminal Mischief"]
Theft_from_auto_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Theft from Vehicle"]
Narcotics_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Narcotic / Drug Law Violations"]
Fraud_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Fraud"]
Agg_Assault_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Aggravated Assault No Firearm"]
Burglary_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Burglary Residential"]
Robbery_df = philly_clean_df.loc[philly_clean_df["Offense"] == "Robbery No Firearm"]

# Top 10 Crime data for pie chart
Total = philly_clean_df["Offense"].count()
AllOther = AllOther_df["Offense"].count()
Thefts = Thefts_df["Offense"].count()
Other_Assaults = Other_Assaults_df["Offense"].count()
Vandalism = Vandalism_df["Offense"].count()
Theft_from_auto = Theft_from_auto_df["Offense"].count()
Narcotics = Narcotics_df["Offense"].count()
Fraud = Fraud_df["Offense"].count()
Agg_Assault = Agg_Assault_df["Offense"].count()
Burglary = Burglary_df["Offense"].count()
Robbery = Robbery_df["Offense"].count()
OtherOther = (Total - AllOther - Thefts - Other_Assaults - Vandalism - Theft_from_auto - Narcotics - Fraud - 
              Agg_Assault - Burglary - Robbery)

print(non_violent_crimes, violent_crimes)
print(AllOther, Thefts, Other_Assaults, Vandalism, Theft_from_auto, Narcotics, Fraud, Agg_Assault, Burglary, Robbery)

OtherOther
Total
```

```python
# Plot pie chart of crime types

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
plt.suptitle("Philadelphia Crime Summary")

# First Plot
values1 = [non_violent_crimes, violent_crimes]
labels1 = ["Non-Violent", "Violent"]
ax1.pie(values1, labels=labels1, autopct="%1.1f%%", shadow=False, startangle=90)

# Second Plot
values2 = [AllOther, Thefts, Other_Assaults, Vandalism, Theft_from_auto, Narcotics, Fraud, Agg_Assault, Burglary, Robbery, OtherOther]
labels2 = ["All Other Offenses", "Thefts", "Other Assaults", "Vandalism/Criminal Mischief", "Theft from Vehicle", 
           "Narcotic / Drug Law Violations", "Fraud", "Aggravated Assault No Firearm", "Burglary Residential", "Robbery No Firearm", 
           "Other Remaining Major Crimes"]
ax2.pie(values2, labels=labels2, autopct="%1.1f%%", shadow=False, startangle=180)

# Plot
plt.subplots_adjust(wspace=.5)
plt.savefig("Output/phillypie.png", dpi=300)   
plt.show()
```

```python
# Group data into Violent / Non-violent categories, trends by month 
philly_trend_df = philly_clean_df.groupby(["Is Violent", "Year", "Month"]).count()

# philly_trend_df_df = philly_trend_df.sort_values(by=["Year", "Month"], ascending=True)dddd
philly_trend_df.head()
```

```python
# Break trend data into violent and non-violent data frames and sort by Year and Month
# Nonviolent data
nonviolent_df = philly_trend_df.loc["Non-Violent"]
nonviolent_df = nonviolent_df.sort_values(by=["Year", "Month"], ascending=True)

# Violent data
violent_df = philly_trend_df.loc["Violent"]
violent_df = violent_df.sort_values(by=["Year", "Month"], ascending=True)
```

```python
# Generate line plot as figure
# Step 1: Configure chart, xticks, and labels/titles
plt.figure(figsize=(12, 5))
plt.xticks(np.arange(16), ["Mar-19", "Apr-19", "May-19", "June-19", "Jul-19", "Aug-19", "Sep-19", "Oct-19", 
                      "Nov-19", "Dec-19", "Jan-20", "Feb-20", "Mar-20", "Apr-20", "May-20", "Jun-20"])
plt.title("Philadelphia City Monthly Crime Rates -- Violent vs. Non-Violent")
plt.xlabel("Month")
plt.ylabel("# Reported Cases")

# Step 2:  Configure data and plot
plt.plot(list(nonviolent_df["Offense"]), label="Non-Violent Crimes")
plt.plot(list(violent_df["Offense"]), label="Violent Crimes")
plt.legend(loc="upper right")
plt.savefig("Output/phill_trend.png", dpi=300) 
plt.show()
```

```python
# Summary look at Violent crimes:
Philly_Vio_Sum_df = philly_clean_df.loc[philly_df["Is Violent"] == "Violent"]

Philly_Vio_Sum_df.info()
```

```python
# Violent crime trends
Vio_Trend_df = Philly_Vio_Sum_df.groupby(["Offense", "Year", "Month"]).count()
Vio_Trend_df.head()
```

```python
# Create violent crime plot data
robbery_df = Vio_Trend_df.loc["Robbery No Firearm"]
AggAssault_df = Vio_Trend_df.loc["Aggravated Assault No Firearm"]
OtherAssaults_df = Vio_Trend_df.loc["Other Assaults"]
Assault_Firearm_df = Vio_Trend_df.loc["Aggravated Assault Firearm"]

robbery = robbery_df["Is Violent"]
AggAssault = AggAssault_df["Is Violent"]
OtherAssaults = OtherAssaults_df["Is Violent"]
Assault_Firearm = Assault_Firearm_df["Is Violent"]
```

```python
# Plot violent crime trends
# Step 1: Configure chart, xticks, and labels/titles
plt.figure(figsize=(12, 5))
plt.xticks(np.arange(16), ["Mar-19", "Apr-19", "May-19", "June-19", "Jul-19", "Aug-19", "Sep-19", "Oct-19", 
                      "Nov-19", "Dec-19", "Jan-20", "Feb-20", "Mar-20", "Apr-20", "May-20", "Jun-20"])
plt.title("Philadelphia City Violent Crime Trends")
plt.xlabel("Month")
plt.ylabel("# Reported Cases")

# Step 2:  Configure data and plot
plt.plot(list(AggAssault), label="Aggravated Assault No Firearm")
plt.plot(list(OtherAssaults), label="Other Assaults")
plt.plot(list(robbery), label="Robbery No Firearm")
plt.plot(list(Assault_Firearm), label="Aggravated Assault Firearm")
plt.legend(loc="center left")
plt.savefig("Output/philly_violent.png", dpi=300) 
plt.show()
```
